```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### -Описание датасета

- **id** - Уникальный ID для каждого дома
- **date** - Дата продажи дома
- **price** - Стоимость продажи дома
- **bedrooms** - Кол-во спален
- **bathrooms** - Кол-во ванных комнат (0.5 - туалет без душа)
- **sqft_living** - Кв. метры жилые
- **sqft_lot** - Кв. метры общие
- **floors** - Кол-во этажей
- **waterfront** - Есть набержная или нет

- **view** - Значение от 0 до 4 насколько хороший вид
- **condition** - Значение от 1 до 5 насколько хорошее состояние
- **grade** - Значение от 1 до 13, где 1-3 плохая конструкция здания и дизайн, 7 - средний уровень конструкции и дизайна, 11-13 - высокое качество конструкции и дизайна
- **sqft_above** - Кв. метры дома, которые находятся выше земли
- **sqft_basement** - Кв. метры дома, которые находятся ниже земли
- **yr_built** - Год постройки дома
- **yr_renovated** - Год ремонта дома

- **zipcode** - Индекс
- **lat** - Широтма
- **long** - Долгота
- **sqft_living15** - Кв. метры жилой площади у 15 соседей
- **sqft_lot15** - Кв. метры общей площади у 15 соседей

- Посмотреть на описание признаков и на их содержание

Датасет показывает основную тенденцию ценообразования дома с точки зрения его местоположения, района постройки, интерьера и т. д.
Есть индекс записи, id записи, дата,цена, кол-во спален, жилая площадь и т.д.

### -Изучить стоимость недвижимости

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['price'])
plt.title('Стоимость недвижимости')
plt.xlabel('Стоимость, у.е.')
plt.ylabel('Количество домов, шт.')
```

### -Изучить распределение жилой квадратуры

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['sqft_living'])
plt.title('Жилая площадь, кв.м')
plt.xlabel('Площадь, кв.м.')
plt.ylabel('Количество домов, шт.')
```

### -Изучить рапределение года постройки

```python
counts = df['yr_built'].value_counts()
plt.scatter(x=counts.index, y=counts.values)
plt.title('Распределение застройки')
plt.xlabel('Год постройки')
plt.ylabel('Количество домов, шт.')
```

### -Изучите распределение этажей домов

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['floors'])
plt.title('Распределение этажей')
plt.xlabel('Кол-во этажей, шт.')
plt.ylabel('Количество домов, шт.')
```

### -Изучите распределение состояния домов

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['condition'])
plt.title('Распределение домов по состоянию (5-ти бальная система)')
plt.xlabel('Состояние')
plt.ylabel('Количество домов, шт.')
```
